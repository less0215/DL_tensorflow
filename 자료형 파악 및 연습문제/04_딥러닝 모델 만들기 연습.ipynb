```python
import tensorflow as tf

# 모델 만들기 1단계: 데이터 준비
### 입·출력 값 정하기
train_x = [1, 2, 3, 4, 5, 6, 7]
train_y = [3, 5, 7, 9, 11, 13, 15]
# y는 x에 2를 곱하고 1을 더해야 나온다.
# 우리는 tf를 이용해서 위 결과 값이 나오는지 증명해야 한다.

# 모델 만들기 2단계: 모델 파라미터 초기화
### weight 값 정하기
w1 = tf.Variable(0.1)
w2 = tf.Variable(0.1)

# 모델 만들기 3단계: 모델 구조 및 손실 함수 정의
### 손실함수 만들기
    # 만드는 이유는 예측값과 실제값의 차이를 줄이기 위해서
def 손실함수(w1, w2):
    # 예측_y를 손실함수 내에서 사용하는 이유
        # 밖에서 예측_y를 정의하면, 최초 값만 계속 사용한다.
        #     그래서 제대로 된 학습이 이루어지지 않는다.
    # 예측 값 식은 정해진 게 없다. 간단한 일차함수부터 시작해서 복잡한 함수까지 사용해 본다.
    예측_y = train_x*w1 + w2  # 여기서 모델 구조가 정의됨
    # 실제 Mean Squared Error와는 입력 순서도 다르고, 제곱도 없다.
        # 그러나 이는 tf가 정의한 입력 방식이다.
        #     mse(실제값, 에측값)으로 적어도 정상적으로 Mean Squared Error가 작동한다.
    return tf.keras.losses.mse(train_y, 예측_y)

'''
# Mean Squared Error 구하는 정확한 수식
def mse(y_true, y_pred):
    오차 = y_pred - y_true
    제곱오차 = 오차 ** 2
    평균제곱오차 = tf.reduce_mean(제곱오차)
    return 평균제곱오차
'''

# 모델 만들기 4단계: 최적화 알고리즘 선택
### optimizer(최적화 알고리즘) 만들기
    # 최적화 알고리즘을 만드는 이유는 예측값과 실제값의 오차를 줄이기 위해서.
opt = tf.keras.optimizers.Adam(learning_rate=0.01)

# 모델 만들기 5단계: 학습 과정 정의
### 경사하강법 방법 정하기
# opt.minimize(손실함수(a, b), var_list=[w1, w2])
# lambda를 사용하는 이유
    # minimize() 인자로는 함수명만 올 수 있기 때문이다.
    #     lambda를 이용해서 인자를 포함한 손실함수를 하나의 함수명으로 만들어준다.
opt.minimize(lambda:손실함수(w1, w2), var_list=[w1, w2])

# 모델 만들기 6단계: 모델 학습 실행
### 경사하강법 실행하기
for i in range(2900):
    opt.minimize(lambda:손실함수(w1, w2), var_list=[w1, w2])
    print(w1.numpy(), w2.numpy())

# 모델 만들기 7단계: 모델 검증
검증 = 1*0.2 + 0.2

print(검증)
```
