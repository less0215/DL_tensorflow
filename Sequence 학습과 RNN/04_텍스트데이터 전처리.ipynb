정답 데이터(레이블) 만들기

```python
'''
.where(조건, True일때 값, False일때 값)
  해당 코드는 3점 미만인 경우 0, 3점 초과인 경우 1을 반환하는 새로운 열을 추가하는 코드다.
raw['label'] = np.where(raw['rating'] > 3, 1, 0)


'''


import pandas as pd
import numpy as np

# raw = pd.read_table('shopping.txt')
raw = pd.read_table('shopping.txt', names = ['rating', 'review'])
# print(raw)

raw['label'] = np.where(raw['rating'] > 3, 1, 0)

print(raw)
```

텍스트데이터 전처리

```python
'''
# 정규 표현식을 이용해 한글·숫자·공백을 모두 제거
raw['review'] = raw['review'].str.replace('[^ㄱ-ㅎㅏ-ㅣ가-힣0-9 ]', '')

# review 열 외 다른 열에 공백 확인
# raw.isnull().sum()

# review에 있는 중복 행 제거 (100% 일치하는 행을 제거 'Hello', 'Hello '는 다른 행임.)
  # subset = ['review']
    raw라는 큰 집합 내에 있는 하위집합(subset)인 review내에 있는 걸 찾아서 제거해라
  # inplace = True
    수정 내용을 원본 데이터에 반영해라
raw.drop_duplicates(subset = ['review'], inplace = True)

# review 내에 있는 내용 리스트로 변환
유니크문자 = raw['review'].tolist()
# 유니크문자 내에 있는 문자열 공백 없이 모두 합치기
  이는 '문자 단위'로 데이터를 분석하기 위해서 필요하다.
    만약 '단어 단위'라면, 공백을 함부로 제거하지 않는다.
유니크문자 = ''.join(유니크문자)
# 유니크한 문자를 선별 후 리스트에 담는다.
유니크문자 = list(set(유니크문자))
# 문자를 오름차순으로 정렬
유니크문자.sort()
print(유니크문자[0:100])

'''


import pandas as pd
import numpy as np

# raw = pd.read_table('shopping.txt')
raw = pd.read_table('shopping.txt', names = ['rating', 'review'])
# print(raw)

raw['label'] = np.where(raw['rating'] > 3, 1, 0)
# print(raw)

raw['review'] = raw['review'].str.replace('[^ㄱ-ㅎㅏ-ㅣ가-힣0-9 ]', '')

# raw.isnull().sum()

raw.drop_duplicates(subset = ['review'], inplace = True)

유니크문자 = raw['review'].tolist()
유니크문자 = ''.join(유니크문자)
유니크문자 = list(set(유니크문자))
유니크문자.sort()
print(유니크문자[0:100])
```
