```python
import tensorflow as tf

Pmodel = tf.keras.models.load_model('/content/drive/MyDrive/model1')
```

텍스트 to 숫자

```python
text = open('/content/drive/MyDrive/pianoabc.txt', 'r').read()
# print(text)

### 중복 없이 텍스트 리스트로 분류하기

유니크text = list(set(text))
유니크text.sort
# print(유니크text)

### 문자를 숫자로 바꾸기
### 숫자를 문자로 바꾸기

text_to_num = {}
num_to_text = {}

for i, data in enumerate(유니크text):
  text_to_num[data] = i
  num_to_text[i] = data

### 원본데이터(텍스트)를 모두 숫자로 바꾸기

변환결과 = []

for i in text:
  변환결과.append(text_to_num[i])

첫입력값 = 변환결과[117: 117+25]
첫입력값 = tf.one_hot(첫입력값, 31)
print(첫입력값)

# 첫입력값.shape
```
