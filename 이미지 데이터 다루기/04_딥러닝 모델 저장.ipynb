모델 전체 저장하기

```python
'''
model.save('새폴더/model1.keras')

save('파일경로')를 사용해서 모델을 저장한다.

.keras 확장자를 가진 파일로 저장된다.
'''

import tensorflow as tf
import numpy as np

(trainX, trainY), (testX, testY) = tf.keras.datasets.fashion_mnist.load_data()

trainX = trainX / 255.0
testX = testX / 255.0

trainX = trainX.reshape( (trainX.shape[0], 28,28,1) )
testX = testX.reshape( (testX.shape[0], 28,28,1) )

model = tf.keras.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax'),
])

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['acc'])
model.fit(trainX, trainY, validation_data=(testX, testY), epochs=3)
```

모델 W값만 저장하기 (epoch가 실행 될 때마다 각 w값 저장하기)

```python
'''
콜백함수 = tf.keras.callbacks.ModelCheckpoint(
    filepath='체크포인트/mnist{epoch}.weights.h5',
    save_weights_only=True,
    save_freq = 'epoch'
)

위 코드를 작성해서 w값을 순간순간 저장한다.

filepath='파일경로' // 파일경로를 작성할 때, w값만 저장한다면 .weights.h5'를 작성해야 한다.

# w값만 저장하겠다는 의미
save_weights_only=True,
# w값 저장을 언제? epoch 할 때마다.
save_freq = 'epoch'

'''


import tensorflow as tf
import numpy as np

(trainX, trainY), (testX, testY) = tf.keras.datasets.fashion_mnist.load_data()

trainX = trainX / 255.0
testX = testX / 255.0

trainX = trainX.reshape( (trainX.shape[0], 28,28,1) )
testX = testX.reshape( (testX.shape[0], 28,28,1) )

model = tf.keras.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax'),
])

콜백함수 = tf.keras.callbacks.ModelCheckpoint(
    filepath='체크포인트/mnist{epoch}.weights.h5',
    save_weights_only=True,
    save_freq = 'epoch'
)

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['acc'])
model.fit(trainX, trainY, validation_data=(testX, testY), epochs=3, callbacks=[콜백함수])
```

모델 W값만 저장하기 (가장 높은 accuracy 값만 저장하기)

```python
'''
# 벨리데이션 정확도를 모니터링 하겠다는 의미
monitor = 'val_acc',
# 모니터링 해서 가장 높은 값만 저장하겠다는 의미
mode = "max",


'''


import tensorflow as tf
import numpy as np

(trainX, trainY), (testX, testY) = tf.keras.datasets.fashion_mnist.load_data()

trainX = trainX / 255.0
testX = testX / 255.0

trainX = trainX.reshape( (trainX.shape[0], 28,28,1) )
testX = testX.reshape( (testX.shape[0], 28,28,1) )

model = tf.keras.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax'),
])

콜백함수 = tf.keras.callbacks.ModelCheckpoint(
    filepath='체크포인트/mnist.weights.h5',
    monitor = 'val_acc',
    mode = "max",
    save_weights_only=True,
    save_freq = 'epoch'
)

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['acc'])
model.fit(trainX, trainY, validation_data=(testX, testY), epochs=3, callbacks=[콜백함수])
```
